## Decode base64 encoded text from the LegiScan API

```python
import pandas as pd
# load the bill_texts_df dataframe from json file
bill_texts_df = pd.read_json('trans_bill_texts.json')
```

```python
bill_texts_df['text'].iloc[0]
```

```python
# loop through the bill_texts_df dataframe and get a count of how many docs are of each mime_type
mime_types = {}
for i in range(len(bill_texts_df)):
    mime_type = bill_texts_df['text'].iloc[i]['mime_id']
    if mime_type in mime_types:
        mime_types[mime_type] += 1
    else:
        mime_types[mime_type] = 1
```

```python
# print the counts of each mime_type. mime_type 2 is pdf, mime_type 1 is html
print(mime_types)
```

```python
# loop through the bill_texts_df dataframe and get the 'doc' from the dict in the 'text' column and decode it from base64 and save it as a file
import base64
for i in range(len(bill_texts_df))[:15]:
    doc_id = bill_texts_df['text'].iloc[i]['doc_id']
    mime_id = bill_texts_df['text'].iloc[i]['mime_id']
    encoded_doc = bill_texts_df['text'].iloc[i]['doc']
    decoded_doc = base64.b64decode(encoded_doc)
    # if mime_id == 2, save the doc as a pdf; if mime_id == 1 save as html file
    if mime_id == 2:
        with open("bill" + str(doc_id) + ".pdf", "wb") as file:
            file.write(decoded_doc)
    elif mime_id == 1:
        with open("bill" + str(doc_id) + ".html", "wb") as file:
            file.write(decoded_doc)    
```
